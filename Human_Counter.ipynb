```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
def detect(img_p):
  f_casc = cv2.CascadeClassifier('/content/haarcascade_frontalface.xml')
  e_casc = cv2.CascadeClassifier('/content/haarcascade_eye.xml')
  img = cv2.imread(img_p)
  gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
  face = f_casc.detectMultiScale(gray)
  eye = e_casc.detectMultiScale(gray)
  for (x,y,w,h) in face:
    cv2.rectangle(img, (x,y), (x+w,y+h), (100,10,250),2)
  for (x,y,w,h) in eye:
    cv2.rectangle(img, (x,y), (x+w, y+h), (200,200,0), 2)
  cv2_imshow(img)
```

```python
detect("/content/Sample1.jpg")
```
